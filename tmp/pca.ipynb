```python
from pandas import read_csv
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
from  sklearn.decomposition import PCA
```

```python
df = read_csv('data\\housing_data_na.csv')
df.head()
```

```python
df.dropna(inplace=True)
df_X = df.drop(columns=['MEDV'])
```

## Gia tri rieng cua ma tran,  Vector rieng

```python
pca = PCA(n_components=13)
pca.fit(df_X)
```

```python
# gia tri ring
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ / sum (pca.explained_variance_)
```

```python
# vector rieng (5 vetor)
pca.components_
```

```python
# giam chiec vector df_X
X_T = pca.transform(df_X)
X_T.shape
```

```python
np.corrcoef(X_T.T)
```

```python
y = df['MEDV'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_T,y)
```

```python
lg = LinearRegression()
lg.fit(X_train, y_train)
```

```python
lg.coef_, lg.intercept_
```

```python
lg.score(X_train, y_train), lg.score(X_test, y_test)
```
